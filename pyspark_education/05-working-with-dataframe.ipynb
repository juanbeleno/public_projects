````
from pyspark.sql.functions import to_date, to_timestamp, round, col, expr
````

````
# Loading a DataFrame using the format method
# This is preferred because this is the standard format to load data from any format
# Actions: kick off a spark job, execute, and return to the spark driver
raw_fire_df = spark.read \
            .format("csv") \
            .option("header", "true") \
            .option("inferSchema", "true") \
            .load("/databricks-datasets/learning-spark-v2/sf-fire/sf-fire-calls.csv")
````

````
# There are two problems with this DataFrame:
# 1. Column names are not standardized (Whitespaces)
# 2. Data fields are of string type
display(raw_fire_df.limit(10))
````

````
# Transformation: modify the DataFrame without activating a spark job. The input and output is a DataFrame -- e.g., withColumnRenamed
# Spark DataFrames are immutable, so we need to transform an existing DataFrame and asign it to a new DataFrame
renamed_fire_df = raw_fire_df \
                    .withColumnRenamed("Call Number", "CallNumber") \
                    .withColumnRenamed("Unit ID", "UnitID") \
                    .withColumnRenamed("Incident Number", "IncidentNumber") \
                    .withColumnRenamed("Call Date", "CallDate") \
                    .withColumnRenamed("Watch Date", "WatchDate") \
                    .withColumnRenamed("Call Final Disposition", "CallFinalDisposition") \
                    .withColumnRenamed("Available DtTm", "AvailableDtTm") \
                    .withColumnRenamed("Zipcode of Incident", "Zipcode") \
                    .withColumnRenamed("Station Area", "StationArea") \
                    .withColumnRenamed("Final Priority", "FinalPriority") \
                    .withColumnRenamed("ALS Unit", "ALSUnit") \
                    .withColumnRenamed("Call Type Group", "CallTypeGroup") \
                    .withColumnRenamed("Unit sequence in call dispatch", "UnitSequenceInCallDispatch") \
                    .withColumnRenamed("Fire Prevention District", "FirePreventionDistrict") \
                    .withColumnRenamed("Supervisor District", "SupervisorDistrict")
````

````
display(renamed_fire_df.limit(10))
````

````
# Utility Methods: They don't transform a DataFrame nor activate a spark job
renamed_fire_df.printSchema()
````

````
# withColumn is a common transformation
fire_df = renamed_fire_df \
            .withColumn("AvailableDtTm", to_timestamp("AvailableDtTm", "MM/dd/yyyy hh:mm:ss a")) \
            .withColumn("Delay", round(col("Delay"), 2))
````

````
display(fire_df.limit(10))
````

````
fire_df.printSchema()
````

````
# We want to cache the DataFrame because we are going to use it heavely for data analysis
fire_df.cache()
````

##### Q1. How many distinct types of call were made to the Fire Departament?

```
SELECT COUNT(DISTINCT CallType) AS num_distinct_call_types
FROM demo_db.fire_service_calls_tbl
WHERE CallType IS NOT NULL
```

````
# First approach
# 1. Create a view of the dataframe
# 2. Run SQL on the view
fire_df.createOrReplaceTempView("fire_services_calls_view")
q1_sql_df = spark.sql(
    """
    SELECT COUNT(DISTINCT CallType) AS num_distinct_call_types
    FROM fire_services_calls_view
    WHERE CallType IS NOT NULL
    """
)
display(q1_sql_df.limit(10))
````

````
# Second approach: Chain of DataFrame transformations
# 1. Filter the records and call those where CallType is not null
# 2. Select the CallType column
# 3. Take only distinct call typess
# 4. Show the count

# Keep actions, transformation, and utility methods in separated lines
# because we cannot chain more methods after an action

# 3 transformation: where, select, distinct
q1_df = fire_df \
            .where("CallType IS NOT NULL") \
            .select("CallType") \
            .distinct()
# 1 action method: count that kick off a spark job
print(q1_df.count())
````

##### Q2. What were the distict types of calls made to the Fire Departament?

```
SELECT DISTINCT CallType
FROM demo_db.fire_service_calls_tbl
WHERE CallType IS NOT NULL
```

````
q2_df = fire_df \
            .where("CallType IS NOT NULL") \
            .select(expr("CallType AS distinct_call_type")) \
            .distinct()
q2_df.limit(100).show()
# display(q2_df.limit(100))
````

##### Q3. Find all the call identifiers for delayed times greather than 5 minutes?

```
SELECT CallNumber, Delay
FROM demo_db.fire_service_calls_tbl
WHERE Delay > 5
LIMIT 100
```

````
q3_df = fire_df \
            .where("Delay > 5") \
            .select("CallNumber", "Delay") \
            .distinct() \
            .limit(100) \
            .show()
````

##### Q4. What were the most common call types?

```
SELECT CallType, COUNT(*) AS num_calls
FROM demo_db.fire_service_calls_tbl
GROUP BY CallType
ORDER BY num_calls DESC
LIMIT 100
```

````
# If count() goes after groupBy(), then it's a tranformation instead of an action
# DataFrame.count() -> Action
# GroupedData.count() -> Transformation

# select and where can swipe their positions without affecting the query result

q3_df = fire_df \
            .select("CallType") \
            .where("CallType IS NOT NULL") \
            .groupBy("CallType") \
            .count() \
            .orderBy("count", ascending=False) \
            .limit(100) \
            .show()
````
