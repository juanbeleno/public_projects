There are 4 possible categorical variables on the dataset: currency, listing_type, shipping_logistic_type, shipping_payment. What are the possible values for those categorical variables?

```python
categorical_columns = ['currency', 'listing_type', 'shipping_logistic_type', 'shipping_payment']
for column in categorical_columns:
    print(f'Column: {column}')
    print(train_data[column].value_counts())
    print('*******************************')
```

This data seems to belong to 3 different countries: Brazil, Mexico and Argentina. So I think, I should create models for the 3 countries. We need to know the distributions between two categories to try to find patterns.

```python
for column1 in categorical_columns:
    for column2 in categorical_columns:
        if column1 != column2:
            print(pd.crosstab(train_data[column1], train_data[column2]))
            print('*******************************')
```

I want to know the impact of each value of the categorical variables on the sold_quantity variable.

```python
categorical_columns = ['currency', 'listing_type', 'shipping_logistic_type', 'shipping_payment']
for column in categorical_columns:
    print(f'Column: {column}')
    values = train_data[column].unique().tolist()
    for value in values:
        subset = train_data[train_data[column] == value].copy()
        print(f'{value}: {subset["sold_quantity"].mean()}')
    print('*******************************')
```

## Hipothesis: The sold_quantity changes depending on the day of the week and day of the month

```python
from datetime import datetime

start_date = datetime(2021, 2, 1)
unique_dates = train_data['date'].unique().tolist()
dates_dict = {x: (datetime.strptime(x, '%Y-%m-%d') - start_date).days for x in unique_dates}
train_data['day_index'] = train_data['date'].apply(lambda x: dates_dict[x])

train_data['date'] = pd.to_datetime(train_data['date'])

# 0 = Monday and 6 = Sunday
train_data['day_of_week'] = train_data['date'].apply(lambda x: x.dayofweek)
train_data['day_of_month'] = train_data['date'].apply(lambda x: x.day)
```

```python
train_data.pivot_table(index='currency', columns='day_of_week', values='sold_quantity',
                aggfunc='sum', fill_value=0)
```

```python
train_data.pivot_table(index='currency', columns='day_of_month', values='sold_quantity',
                aggfunc='sum', fill_value=0)
```

It seems to me that at the end of the month there are less sold_quantities and people doesn't buy products on weekends.

# Analysis of 6 random skus

```python
import matplotlib.pyplot as plt

random_skus = train_data['sku'].sample(6).tolist()
random_skus

fig, ax = plt.subplots(3, 2, figsize=(15, 15))
index = 0

for i in range(3):
    for j in range(2):
        sku = random_skus[index]
        subset = train_data[train_data['sku'] == sku].copy()
        #print(subset)
        ax[i, j].bar(
            subset.day_index,
            subset.sold_quantity,
            color="k",
            alpha=0.6
        )
        ax[i, j].set_title(sku)
        index += 1

plt.show()
```

The patterns on random samples aren't as clear as I though. Sometimes, there are gaps, sometimes sells are concentrated on some days at one part of the time series. So, I'll start by creating a model and perform error analysis on the skus with more error.
